```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_boston
from sklearn.tree import DecisionTreeRegressor
```

```python
boston = load_boston()
data = pd.DataFrame(boston.data,columns=boston.feature_names)
data = pd.concat([data,pd.Series(boston.target,name='MEDV')],axis=1)
data.head()
```

```python
X = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

```python
x_training_set, x_test_set, y_training_set, y_test_set = train_test_split(X,y,test_size=0.10,random_state=42, shuffle=True)
```

```python
model =  DecisionTreeRegressor(max_depth=2,random_state=0)
estimator = model.fit(x_training_set, y_training_set)
```

```python
from sklearn.tree import export_graphviz
# Export as dot file
export_graphviz(estimator, out_file='tree.dot', 
                feature_names = boston.feature_names,
                class_names = boston.target,
                rounded = True, proportion = False, 
                precision = 2, filled = True)
```

```python
from sklearn.tree import _tree 

def get_rules(tree, feature_names):
    rules = list()
    
    def traverse_nodes(node_id=0,
                       operator=None,
                       threshold=None,
                       feature=None,
                       conditions=[]):
        
        if node_id != 0:
            
            if feature_names is not None:
                feature_name = feature_names[feature]
            else:
                feature_name = feature
                
#             rule_condition = RuleCondition(feature_index=feature,
#                                            threshold=threshold,
#                                            operator=operator,
#                                            support = tree.n_node_samples[node_id] / float(tree.n_node_samples[0]),
#                                            feature_name=feature_name)

            rule_condition = (feature_names[feature], np.round(threshold,2), operator)
    
            new_conditions = conditions + [rule_condition]
            
        else:
            new_conditions = []
            
        ## if not terminal node
        if tree.children_left[node_id] != tree.children_right[node_id]: 
            
            feature = tree.feature[node_id]
            threshold = tree.threshold[node_id]
            
            left_node_id = tree.children_left[node_id]
            traverse_nodes(left_node_id, "<=", threshold, feature, new_conditions)
            
            right_node_id = tree.children_right[node_id]
            traverse_nodes(right_node_id, ">", threshold, feature, new_conditions)
            
        else: # a leaf node
            if len(new_conditions)>0:
#                 new_rule = Rule(new_conditions,tree.value[node_id][0][0])
                rules.append(new_conditions)
            else:
                pass #tree only has a root node!
            return None

    traverse_nodes()
    
    return rules
```

```python
rules = get_rules(model.tree_, boston.feature_names)
```

```python
def apply_rule_to_df(df, conditions, col_name):
    satisfy_df = df.copy()
    for condition in conditions:
        if condition[2] == '<=':
            satisfy_df = satisfy_df[satisfy_df[condition[0]] <= condition[1]]
        else:
            satisfy_df = satisfy_df[satisfy_df[condition[0]] > condition[1]]
    index_list = list(satisfy_df.index.values)
    
    new_col = list()
    for i in range(0,len(df)):
        if df.index.values[i] in index_list:
            new_col.append(1)
        else:
            new_col.append(0)
            
    df.loc[:,col_name] = new_col
    return df
```

```python
apply_rule_to_df(data, rules[0],'test_col')
```
